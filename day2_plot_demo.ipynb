```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
loans = pd.read_csv("data/prosperLoanData_truncated.csv")
```

```python
loans.columns
```

```python
data = loans.StatedMonthlyIncome[(loans.StatedMonthlyIncome > 100) & (loans.StatedMonthlyIncome < 1E6)]
plt.figure(figsize=(10, 3))
plt.hist(data, bins=np.arange(0,1E6,500), color="green")
plt.xlabel("Stated Monthly Income")
plt.ylabel("Number of Borrowers")
plt.xscale('log')
plt.show()
```

```python
loans.StatedMonthlyIncome.describe()
```

```python
heights, edges = np.histogram(data, bins=np.arange(0,40000,500) )
heights
```

```python
probs = heights / sum(heights)
plt.figure(figsize=(10, 3))
plt.bar(edges[:-1], probs, width=edges[1] - edges[0], color="green")
plt.show()
```

```python
probs = heights / sum(heights)
plt.figure(figsize=(10, 3))
plt.bar(edges[:-1], probs.cumsum(), width=edges[1] - edges[0], color="green")
plt.plot([0, 40000], [0.5, 0.5])
plt.plot([0, 40000], [0.7, 0.7], color='red')

plt.show()
```
