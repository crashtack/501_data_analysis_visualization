```python
from random import randint, randrange
import matplotlib.pyplot as plt
from statistics import mean
import pandas as pd
```

```python
def best_fit_slope_and_intercept(xs, ys):
    """Return m and b of y = mx + b"""
    m = ((((mean(xs) * mean(ys))) - mean(xs * ys)) /
         ((mean(xs) * mean(xs)) - mean(xs * xs)))

    b = mean(ys) - m * mean(xs)

    return m, b

def line(m, b, x):
    """Generate a line for m x and b data"""
    output = m * x + b
    return output
```

## Create Random Channel Hope data
Hoop over 50 channels in a pseudo random fashon spending equal
time in each channel

```python
def rand_hoop():
    """Create X Y data for channel hooping test"""
    # Select a random channel 1 thru 50
    # if the channel has already been selected select anther
    visited = [0 for i in range(50)]
    channel = []
    next_channel = 0
    while sum(visited) < 50:
        next_channel = randrange(50)
        if visited[next_channel] == 0:
            visited[next_channel] = 1
            channel.append(next_channel)
    return channel
```

```python
y = rand_hoop()
X = [x for x in range(50)]
print(y)

xs = pd.Series(X)
ys = pd.Series(y)

m, b = best_fit_slope_and_intercept(xs, ys)
print("m = {}, b = {}".format(m, b))

fit = line(m, b, xs)
```

```python
plt.plot(xs, fit)
plt.scatter(X, y, edgecolor=None, c='red')
plt.xlim(0,50)
plt.ylim(0,50)
plt.show()
```

```python
X = [x for x in range(50)]
xs = pd.Series(X)
sample_size = 1000
m_set = []

for i in range(sample_size):
    y = rand_hoop()
    ys = pd.Series(y)
    m, b = best_fit_slope_and_intercept(xs, ys)
    m_set.append(m)

X = [x for x in range(sample_size)]
xs = pd.Series(X)
m, b = best_fit_slope_and_intercept(xs, m_set)
print("m = {}, b = {}".format(m, b))
fit = line(m, b, xs)
```

```python
plt.plot(xs, fit, color='green')
plt.scatter(X, m_set,
            edgecolor=None,
            alpha=0.1,
            c='red')
# plt.xlim(0,50)
# plt.ylim(0,50)
plt.show()
```
