# Day 1: Uber Data

````python
import csv
active_vehicles = []
trips = []
uber = []
uber2 = {}
with open('Uber-Jan-Feb-FOIL.csv', 'r') as csvfile:
    data = csv.reader(csvfile, delimiter=',', quotechar='|')
    for i, row in enumerate(data):
        if i > 0:
            active_vehicles.append(int(row[2]))
            trips.append(int(row[3]))
            uber.append({'av': int(row[2]),
                         'trips': int(row[3]),
                         'base': row[0]})
            try:
                uber2[row[1]].setdefault(row[0], {'av': int(row[2]), 'trips': int(row[3])})
                
            except KeyError:
                uber2.setdefault(row[1], {row[0]: {'av': int(row[2]), 'trips': int(row[3])}})
````

````python
total_av = 0
total_trips = 0
count = 0
for i in uber:
    total_av += i['av']
    total_trips += i['trips']
    count += 1
print('total_av: {}, total_trips: {}, count: {}'.format(total_av, total_trips, count))
````

````python
average_av = total_av / count
average_trips = total_trips / count
print("average_av: {}".format(average_av))
print('average_trips: {}'.format(average_trips))
````

````python
average_trips_per_vehicle_per_day = average_trips / average_av
print("Average trips per vehicle per day: {}".format(average_trips_per_vehicle_per_day))
````
